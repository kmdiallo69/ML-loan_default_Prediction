```python
def Get_href(products):
    https="https://fr.openfoodfacts.org"
    hrefs=[]
    liens=[]
    for item in range(len(products)):
        res=products[item].find('a')['href']
        lien=https+res
        liens.append(lien)
        hrefs.append(res)
  
    return(hrefs,liens)
```

```python
hrefs,liens=Get_href(products)
```
